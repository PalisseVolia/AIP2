### Prepare package, parameters, data

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
np.random.seed(42)

lr = 1e-2
tot_iter = 1000

x, y = datasets.make_regression(
        n_samples=500, n_features=1, noise=15, random_state=42)
x = x.flatten()
w = np.random.randn(1)
b = np.random.randn(1)
```

### Training

You must write explanation about this section in your report.  (Task1 and Task2)

```python
for it in range(tot_iter):
    pred = x * w + b
    mse = np.mean((pred - y)**2)
    dw = np.mean((2 * (pred - y)) * x)
    db = np.mean(2 * (pred - y))
    w -= lr * dw
    b -= lr * db

pred = x * w + b
mse = np.mean((pred - y)**2)
print("Mean Square Error {}".format(mse))
print("weight:{}, bias:{}".format(w,b))
```

### Visualize
Hint: use plt.plot() with only two data points

```python
plt.scatter(x, y, label='Data Points')

x_line = np.array([x.min()-0.5, x.max()+0.5])
y_line = w * x_line + b
plt.plot(x_line, y_line, color='red', label='Regression line')

plt.xlabel('Input Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```
